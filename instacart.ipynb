```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df_sliced = pd.read_csv('final_data_instacart_400k.csv')
print(df_sliced.head())


print(df_sliced.columns)
```

```python
sns.countplot(x='department', data=df_sliced)
plt.title('Bar Plot for Department')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.distplot(df_sliced['order_hour_of_day'], kde=True)
plt.title('Histogram with KDE for Order Hour of Day')
plt.show()
```

```python
numerical_cols = ['add_to_cart_order', 'order_number', 'order_hour_of_day', 'days_since_prior_order']
sns.pairplot(df_sliced[numerical_cols])
plt.title('Pair Plot for Numerical Columns')
plt.show()
```

```python
corr = df_sliced[numerical_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Between Numerical Features')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='department', y='order_hour_of_day', data=df_sliced)
plt.title('Box Plot of Order Hour of Day Grouped by Department')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='department', y='order_hour_of_day', data=df_sliced)
plt.title('Violin Plot of Order Hour of Day Grouped by Department')
plt.xticks(rotation=90)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_sliced['add_to_cart_order'], df_sliced['order_number'], df_sliced['order_hour_of_day'])
ax.set_xlabel('Add to Cart Order')
ax.set_ylabel('Order Number')
ax.set_zlabel('Order Hour of Day')
plt.title('3D Plot for Add to Cart Order, Order Number, and Order Hour of Day')
plt.show()
```
